```python
import math

x1 = 5
y1 = 3

x2 = 8
y2 = 11

print(
  f'''
    Question : 
      There were two point:
      - point 1: ({x1}, {y1})
      - point 2: ({x2}, {y2})
    Please get the distance between these two points
  '''
)


x_dist = abs(x2 - x1)
y_dist = abs(y2 - y1)

dist = math.sqrt(x_dist**2 + y_dist**2)

print(
  f'''
    Answer:
      Distance between these two points is: {dist}

  '''
)
```

```python
measurements_1 = [2.3, 3.1, 5.6]
measurements_2 = [1.8, 7.0]
measurements_3 = [4.5, 1.5, 6.4, 3.2]

total_measurements = measurements_1 + measurements_2 + measurements_3

measurements_sum = sum(total_measurements)
measurements_len = len(total_measurements)
measurements_ave = measurements_sum / measurements_len

print(
  f'''
  Summary of these numbers is {measurements_sum}.
  Length of these numbers is {measurements_len}.
  Mean of these numbers is {measurements_ave}.
  '''
)
```

```python
values    = [1,5,8,3,5,9,2,5,1,2,8,9]
target    = []
threshold = 6

for i in values:
  if i < threshold:
    print(i)
    target.append(i)

print('Target: ' + str(target))
```

```python
values    = [1,5,8,3,5,9,2,5,1,2,8,9]
target    = []
threshold = 6

for i in range(len(values)):
  specific_num = values[i]
  
  if specific_num < threshold:
    print(specific_num)
    target.append(specific_num)

print('Target: ' + str(target))
```

```python
def odd_or_even(number):

  if number % 2 == 0:
    print(f'{number} is an even number')
  else:
    print(f'{number} is an odd number')

odd_or_even(21)
```
